```python
import pandas as pd
```

```python
train = pd.read_csv('balanced_nair_train.csv')
valid = pd.read_csv('balanced_nair_valid.csv')
```

```python
train
```

```python
train['name'] = train['name'] + '.jpg'
```

```python
valid['name'] = valid['name'] + '.jpg'
```

```python
train = train.sample(frac=1).reset_index(drop=True)
valid = valid.sample(frac=1).reset_index(drop=True)
```

```python
train.to_csv('t.csv', index=None)
valid.to_csv('v.csv', index=None)
```
