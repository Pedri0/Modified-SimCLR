```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image

#https://setosa.io/ev/image-kernels/
```

```python
@tf.function
def sobel_edges(image):
    #Gets only one image from sobel filter dx or dy

    #Args.
        #image: float tensor of shape (height, width, channels)

    #Returns.
        # sobel filter dx or dy depending on random number
    
    sobel_x = tf.constant([[1, 0, -1], [2, 0, -2], [1, 0, -1]], tf.float32, shape = [3, 3, 1, 1])
    sobel_x = tf.tile(sobel_x, [1, 1, 3, 1])
    strides = [1,1,1,1]
    bobel = tf.nn.depthwise_conv2d(image[None, ...], sobel_x, strides, padding='VALID')
    bobel = tf.clip_by_value(bobel, 0., 1.)
    bobel = bobel[0]
    bobel = tf.image.convert_image_dtype(bobel, tf.uint8)
    
    #sobel_x_filter = tf.reshape(sobel_x, [3, 3, 1, 1])
    #sobel_y_filter = tf.transpose(sobel_x_filter, [1, 0, 2, 3])

    #image = tf.placeholder(tf.float32, shape = [None, None])
    #image = tf.expand_dims(tf.expand_dims(image, 0), 3)

    #filtered_x = tf.nn.conv2d(image, sobel_x_filter, strides=[1,1,1,1], padding='SAME')
    #filtered_y = tf.nn.conv2d(image, sobel_y_filter, strides=[1,1,1,1], padding='SAME')

    return bobel
```

```python
@tf.function
def dss(image):
    #Gets only one image from sobel filter dx or dy

    #Args.
        #image: float tensor of shape (height, width, channels)

    #Returns.
        # sobel filter dx or dy depending on random number
    
    sobel_x = tf.constant([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], tf.float32, shape = [3, 3, 1, 1])
    sobel_x = tf.tile(sobel_x, [1, 1, 3, 1])
    strides = [1,1,1,1]
    bobel = tf.nn.depthwise_conv2d(image[None, ...], sobel_x, strides, padding='VALID')
    bobel = tf.clip_by_value(bobel, 0., 1.)
    bobel = bobel[0]
    bobel = tf.image.convert_image_dtype(bobel, tf.uint8)
    
    #sobel_x_filter = tf.reshape(sobel_x, [3, 3, 1, 1])
    #sobel_y_filter = tf.transpose(sobel_x_filter, [1, 0, 2, 3])

    #image = tf.placeholder(tf.float32, shape = [None, None])
    #image = tf.expand_dims(tf.expand_dims(image, 0), 3)

    #filtered_x = tf.nn.conv2d(image, sobel_x_filter, strides=[1,1,1,1], padding='SAME')
    #filtered_y = tf.nn.conv2d(image, sobel_y_filter, strides=[1,1,1,1], padding='SAME')

    return bobel
```

```python
@tf.function
def sobel_x_t(image):
    #Gets sobel dx filter for an image tensor

    #Args.
        #image: float tensor of shape (height, width, channels)

    #Returns.
        # sobel filter dx or dy depending on random number
    
    sobel_x = tf.constant([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], tf.float32, shape = [3, 3, 1, 1])
    sobel_x = tf.tile(sobel_x, [1, 1, 3, 1])
    
    pad_sizes = tf.constant([[1, 1], [1, 1], [0, 0]], tf.int32)
    padded = tf.pad(image, pad_sizes, mode='REFLECT')
    strides = [1,1,1,1]
    sobel = tf.nn.depthwise_conv2d(padded[None, ...], sobel_x, strides, padding='VALID')
    sobel = tf.clip_by_value(sobel, 0., 1.)
    sobel = sobel[0]
    sobel = tf.image.convert_image_dtype(sobel, tf.uint8)
    
    return sobel
```

```python
@tf.function
def sobel_y_t(image):
    #Gets sobel filter dy for an image tensor

    #Args.
        #image: float tensor of shape (height, width, channels)

    #Returns.
        # sobel filter dy
    
    sobel_y = tf.constant([[1, 0, -1], [2, 0, -2], [1, 0, -1]], tf.float32, shape = [3, 3, 1, 1])
    sobel_y = tf.tile(sobel_y, [1, 1, 3, 1])
    
    pad_sizes = tf.constant([[1, 1], [1, 1], [0, 0]], tf.int32)
    padded = tf.pad(image, pad_sizes, mode='REFLECT')
    
    
    strides = [1,1,1,1]
    sobel = tf.nn.depthwise_conv2d(padded[None, ...], sobel_y, strides, padding='VALID')
    sobel = tf.clip_by_value(sobel, 0., 1.)
    sobel = sobel[0]
    sobel = tf.image.convert_image_dtype(sobel, tf.uint8)

    return sobel
```

```python
gg = sobel_x_t(img)
gg = gg.numpy()
plt.figure(figsize = (20,20))
plt.imshow(gg)
```

```python
y = sobel_y_t(img)
y = y.numpy()
plt.figure(figsize = (20,20))
plt.imshow(y)
```
